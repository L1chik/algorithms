```python
# Предобработка

# Knuth Morris Prath
def KMP(register=False):
    if (register == False):
        print('Введите изначальную строку:')
        line = input().lower()
        print('Введите подстроку, которую необходимо найти:')
        pattern = input().lower()
        
    else:
        print('Введите изначальную строку:')
        line = input()
        print('Введите подстроку, которую необходимо найти:')
        pattern = input()
        
    
    pattern_len = len(pattern)
    line_len = len(line)
    lps = [0] * pattern_len
    j = 0
    
    
    def computeLPSArray(pattern, pattern_len, lps):
        length = 0
        lps[0]
        i = 1
        
        while i < pattern_len:
            if pattern[i] == pattern[length]:
                length += 1
                lps[i] = length
                i += 1
                
            else:
                if length != 0:
                    length = lps[length-1]
                else:
                    lps[i] = 0
                    i += 1
                    
                    
    computeLPSArray(pattern, pattern_len, lps)
    i = 0
    
    while i < line_len:
        if pattern[j] == line[i]:
            i += 1
            j += 1
            
        if j == pattern_len:
            print("Индекс начала подстроки: {}".format(str(i-j)))
            j = lps[j-1]
            
        elif i < line_len and pattern[j] != line[i]:
            if j !=0:
                j = lps[j-1]
            else:
                i += 1
                
                
# Boyer Moore

def BM(register=False):
    if (register == False):
        print('Введите изначальную строку:')
        line = input().lower()
        print('Введите подстроку, которую необходимо найти:')
        pattern = input().lower()
        
    else:
        print('Введите изначальную строку:')
        line = input()
        print('Введите подстроку, которую необходимо найти:')
        pattern = input()
        
    
    N_OF_CHARS = 256
    
    def badCharHeuristic(string, size):
        badchar = [-1]*N_OF_CHARS
        
        for i in range(size):
            badchar[ord(string[i])] = i
            
        return badchar
    
    
    def search(line, pattern):
        pattern_len = len(pattern)
        line_len = len(line)
        badchar = badCharHeuristic(pattern, pattern_len)
        s = 0
        
        while(s <= line_len-pattern_len):
            j = pattern_len-1
            
            while j>=0 and pattern[j] == line[s+j]:
                j -= 1
            
            if j<0:
                print("Индекс начала подстроки: {}".format(s))
                s += (pattern_len-badchar[ord(line[s+pattern_len])] if s+pattern_len < line_len else 1)
                
            else:
                s += max(1, j-badchar[ord(line[s+j])])
                
    return search(line, pattern)
```
