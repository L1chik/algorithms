### Сортировка строк матриц

```python
# Предобработка

## QUICKSORT
def quick_sorted(matrix_str):
    if len(matrix_str) <= 1:
        return matrix_str
    else:
        med = statistics.median(matrix_str)
        lower_nums = [j for j in matrix_str if j < med]
        equal_nums = [j for j in matrix_str if j == med]
        bigger_nums = [j for j in matrix_str if j > med]
        
    return quick_sorted(lower_nums) + equal_nums + quick_sorted(bigger_nums)


def quick_sort(matrix):
    for it in range(len(matrix)):
        matrix.iloc[it] = quick_sorted(matrix.iloc[it])
        
    return matrix


## BUBBLESORT
def bubble_sorted(matrix_str):
    length = len(matrix_str)
    
    for it in range(length):
        changes = False
        for j in range(0, length-it-1):
            if matrix_str[j] > matrix_str[j+1]:
                matrix_str[j], matrix_str[j+1] = matrix_str[j+1], matrix_str[j]
                changes = True
                
        if changes == False:
            return matrix_str
            
            
def bubble_sort(matrix):
    for it in range(len(matrix)):
        matrix.iloc[it] = bubble_sorted(matrix.iloc[it])
        
    return matrix
        
        
## SELECTIONSORT
def selection_sorted(matrix_str):
    length = len(matrix_str)
    
    for it in range(length):
        index = it
        for j in range(it+1, length):
            if matrix_str[j] < matrix_str[index]:
                index = j
        matrix_str[it], matrix_str[index] = matrix_str[index], matrix_str[it]
    return matrix_str
            

def selection_sort(matrix):
    for it in range(len(matrix)):
            matrix.iloc[it] = selection_sorted(matrix.iloc[it])   
            
    return matrix
    

## INSERTIONSORT
def insertion_sorted(matrix_str):
    matrix_str = list(matrix_str)
    
    for it, n in enumerate(matrix_str):
        ins_pos = 0
        for j, m in reversed([num for num in enumerate(matrix_str[:it])]):
            if m < n:
                ins_pos = j + 1
                break
        matrix_str.insert(ins_pos, matrix_str.pop(it))
                
    return matrix_str


def insertion_sort(matrix):
    for it in range(len(matrix)):
            matrix.iloc[it] = insertion_sorted(matrix.iloc[it])  
            
    return matrix


## SHELLSORT #non optimized version???
def shell_sorted(matrix_str):
    length = len(matrix_str)
    gap = length 
    
    while gap > 0:
        for it in range(gap, length):
            buff = matrix_str[it]
            j = it
            while j >= gap and matrix_str[j-gap] > buff:
                matrix_str[j] = matrix_str[j-gap]
                j -= gap
            matrix_str[j] = buff
        gap //= 2
    
    return matrix_str


def shell_sort(matrix):
    for it in range(len(matrix)):
            matrix.iloc[it] = shell_sorted(matrix.iloc[it])  
            
    return matrix

## HEAPSORT
def heapify(matrix_str):
    for it in reversed(range(1, len(matrix_str))):
        parent = (it - 1)//2
        if matrix_str[parent] > matrix_str[it]:
            matrix_str[parent], matrix_str[it] = matrix_str[it], matrix_str[parent]
            
            
def heap_sorted(it):
    result = []
    matrix_str = list(it)
    
    while matrix_str:
        heapify(matrix_str)
        result.append(matrix_str.pop(0))
        
    return result


def heap_sort(matrix):
    for it in range(len(matrix)):
        matrix.iloc[it] = heap_sorted(matrix.iloc[it])
        
    return matrix
```

## Задача 2. «Максимальное число»

```python
def max_value(arr):
    empty, vals = [], ""
    length = len(str(max(arr))) + 1
    
    for i in arr:
        buff = str(i) * length
        empty.append((buff[:], i))
        
    empty = list(reversed(heap_sorted(empty)))
    
    for i in empty:
        vals += str(i[1])
        
    if int(vals) == 0:
        return "0"
    
    return vals
```

```python
arr = [3,30,34,5,9]
max_value(arr)
```

## Задача 3. «Сортировка диагоналей в матрице»

```python
 def diag_sort(matrix):
        y, x = len(matrix), len(matrix[0])
        for i in range(2-y, x-1):
            valid = range(max(0, 0-i), min(y, x-i))
            diag, k = heap_sorted([matrix[j][i+j] for j in valid]), 0
            for j in valid:
                matrix[j][i+j], k = diag[k], k + 1
        return matrix
```
